```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates 
from datetime import datetime

Ca = pd.read_csv('all_california_data.csv') #CA data
Te = pd.read_csv('all_texas_data.csv') #Texas data

Te['time'] = pd.to_datetime(Te['time'])

#July in Texas
start_date =  datetime.strptime("2019-07-01 00:00:00 -0500", "%Y-%m-%d %H:%M:%S %z")
end_date =  datetime.strptime("2019-08-01 00:00:00 -0500", "%Y-%m-%d %H:%M:%S %z")

CaJulTime = Te[(Te['time'] >= start_date) & (Te['time'] < end_date)]
CaJulTime = CaJulTime.iloc[::-1]

##Hourly Demand
avgHourlyDemand = np.mean(np.array(CaJulTime['tx_demand']).reshape(31, 24), axis=0)

#Hourly Energy Generation per Source
avgCoal = np.mean(np.array(CaJulTime['tx_coal_gen']).reshape(31, 24), axis=0)
ca_wind_gen = np.mean(np.array(CaJulTime['tx_wind_gen']).reshape(31, 24), axis=0)
ca_solar_gen = np.mean(np.array(CaJulTime['tx_solar_gen']).reshape(31, 24), axis=0)
ca_natural_gen = np.mean(np.array(CaJulTime['tx_natural_gen']).reshape(31, 24), axis=0)
ca_hydro_gen = np.mean(np.array(CaJulTime['tx_hydro_gen']).reshape(31, 24), axis=0)
ca_nuclear_gen = np.mean(np.array(CaJulTime['tx_nuclear_gen']).reshape(31, 24), axis=0)
ca_other_gen = np.mean(np.array(CaJulTime['tx_other_gen']).reshape(31, 24), axis=0)
#TX doesn't have petroleum

#Total Energy Generation
CaJulCoal = CaJulTime['tx_coal_gen']
CaJulWind = CaJulTime['tx_wind_gen']
CaJulSolar = CaJulTime['tx_solar_gen']
CaJulNatural = CaJulTime['tx_natural_gen']
CaJulHydro = CaJulTime['tx_hydro_gen']
CaJulNuclear = CaJulTime['tx_nuclear_gen']
CaJulOther = CaJulTime['tx_other_gen']
totalGen = CaJulCoal + CaJulWind + CaJulSolar + CaJulNatural + CaJulHydro + CaJulNuclear + CaJulOther 
#print(totalGen)
totalGenArr = np.mean(np.array(totalGen).reshape(31, 24), axis=0)

plt.figure(figsize=(10, 7))
plt.plot(range(24), avgHourlyDemand*1e-3, marker='o', linestyle='-',label='Hourly Demand')
plt.plot(range(24), avgCoal*1e-3, marker='o', linestyle='-',label='Coal Generation')
plt.plot(range(24), ca_wind_gen*1e-3, marker='o', linestyle='-',label='Wind Generation')
plt.plot(range(24), ca_solar_gen*1e-3, marker='o', linestyle='-',label='Solar Generation')
plt.plot(range(24), ca_natural_gen*1e-3, marker='o', linestyle='-',label='Natural Gas Generation')
plt.plot(range(24), ca_hydro_gen*1e-3, marker='o', linestyle='-',label='Hydro Generation')
plt.plot(range(24), ca_nuclear_gen*1e-3, marker='o', linestyle='-',label='Nuclear Generation')
plt.plot(range(24), ca_other_gen*1e-3, marker='o', linestyle='-',label='Other Generation')
plt.plot(range(24), totalGenArr*1e-3, marker='o', linestyle='-',label='Total Generation')
plt.title('July 2019 in Texas')
plt.xlabel('Hour')
plt.ylabel('Energy [GWh]')
plt.xticks(range(24))
plt.yticks(range(0, 80, 5 ))
plt.grid(True)
leg = plt.legend( loc = 'upper left')
```
